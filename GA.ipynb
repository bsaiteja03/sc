```python
import numpy as np
import random
```

```python
class Item(object):
    def __init__(self,v,w):
        self.v=v
        self.w=w
ITEMS = [Item(random.randint(0,5),random.randint(0,5)) for x in range (0,5)]
popsize=50
genmax=200
capacity=len(ITEMS)*10
```

```python
def initialize(size):
    pop=[]
    for x in range(0,size):
        pop.append([random.randint(0,1) for i in range(0,len(ITEMS))])
    return pop
```

```python
def fitness(gene):
    total_val=0
    total_weight=0
    idx=0
    for i in gene:
        if idx>len(gene):
            break
        if(i==1):
            total_val+=ITEMS[idx].v
            total_weight+=ITEMS[idx].v
        idx+=1
    if(total_weight>capacity):
        return 0
    else:
        return total_val
```

```python
def mutate(gene):
    pos=random.randint(0,len(gene)-1)
    if(gene[pos]==1):
        gene[pos]=0
    else:
        gene[pos]=1
```

```python
def evovle_pop(pop):
    parent_size=0.2
    mutate_chance=0.08
    parent_lottery=0.05
    no_of_parents=int(parent_size*len(pop))
    parents=pop[:no_of_parents]
    nonparents=pop[no_of_parents:]
    for np in nonparents:
        if parent_lottery >random.random():
            parents.append(np)
    for p in parents:
        if mutate_chance >random.random():
            mutate(p)
    remainingpop=len(pop)-len(parents)
    children=[]
    while(len(children)<remainingpop):
        male=pop[random.randint(0,len(parents)-1)]
        female=pop[random.randint(0,len(parents)-1)]
        crossover_point=int(len(male)/2)
        child=female[crossover_point:]+male[:crossover_point]
        if mutate_chance>random.random():
            mutate(child)
        children.append(child)
    parents.extend(children)
    return parents
```

```python
def main():
    gen=1
    population=initialize(popsize)
    for g in range(0,genmax):
        print("Generation",gen,"with popsize",popsize)
        population=sorted(population,key=lambda x: fitness(x),reverse=True)
        for i in population:        
            print("%s, fit: %s" % (str(i), fitness(i))) 
        population = evovle_pop(population)
        gen+=1
    maxfitness=fitness(population[0])
    print("MAX FITNESS=",maxfitness)
if __name__ == "__main__":
    main()
```
